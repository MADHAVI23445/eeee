```python
df['time'] = df['release_year']/100
df
```

```python
def quality(release_year,time):
    if release_year/time > 10:
        return 'geneerous'
    else:
        return "othrr"
```

```python
df['quality']= df[['release_year','time']].apply(lambda df: quality(df['release_year'],df['time']),axis=1)
df
```

```python
df['quality'] = np.vectorize(quality)(df['release_year'],df['time'])
df
```
